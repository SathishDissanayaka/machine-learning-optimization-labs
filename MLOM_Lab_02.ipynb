```python
#Lab 02
#quiz 01
import pandas as pd
import numpy as np

# Creating the dictionary and DataFrame
data1 = {
    'Name': ['Alice', 'Bob', np.nan, 'David'],
    'Age': [25, np.nan, 22, 28],
    'City': ['Colombo', 'Kandy', 'Galle', np.nan]
}

df1 = pd.DataFrame(data1)

# Show where the null values are
print("Null value matrix (True = missing):")
print(df1.isnull())
print()

# Fill missing values
df1_filled = df1.fillna({
    'Name': 'Unknown',
    'Age': df1['Age'].mean(),
    'City': 'Unknown'
})

print("DataFrame after filling missing values:")
print(df1_filled)
print()

# Drop rows with any missing values
df1_dropped = df1.dropna()

print("DataFrame after dropping rows with missing values:")
print(df1_dropped)

#quiz2
data2 = {
    'ID': [1, 2, 3, 3, 4, 5, 5],
    'Name': ['A', 'B', 'C', 'C', 'D', 'E', 'E']
}
df2 = pd.DataFrame(data2)

print("Original DataFrame:")
print(df2)
print()

# Show duplicated rows (all columns by default)
print("Duplicated rows:")
print(df2[df2.duplicated()])
print()

# Drop duplicates and keep the first occurrence
df2_unique = df2.drop_duplicates()

print("DataFrame after dropping duplicates:")
print(df2_unique)

#quiz 3
data3 = {
    'ID': ['101', '102', '103'],
    'JoinDate': ['2022-01-10', '2022-02-15', '2022-03-01'],
    'Salary': ['50000', '60000', '70000']
}

df3 = pd.DataFrame(data3)

# Check initial data types
print("Initial dtypes:")
print(df3.dtypes)
print()

# Convert data types
df3['ID'] = df3['ID'].astype(int)
df3['JoinDate'] = pd.to_datetime(df3['JoinDate'])
df3['Salary'] = df3['Salary'].astype(float)

# Check data types after conversion
print("Dtypes after conversion:")
print(df3.dtypes)

#quiz 4
data4 = {
    'Product': [' Apple ', 'BaNaNa', ' Cherry', 'DaTE ']
}
df4 = pd.DataFrame(data4)

print("Original DataFrame:")
print(df4)
print()

# Remove leading and trailing spaces
df4['Product'] = df4['Product'].str.strip()
print("After strip():")
print(df4)
print()

# Convert to lowercase
df4['Product'] = df4['Product'].str.lower()
print("After lower():")
print(df4)

#quiz 5
# Original data with outliers
data5 = {
    'Score': [55, 60, 62, 90, 95, 200, 58, 59, 210]
}
df5 = pd.DataFrame(data5)

print("Original DataFrame:")
print(df5)
print()

# Detect outliers (Score > 100)
outliers = df5[df5['Score'] > 100]
print("Outliers:")
print(outliers)
print()

# Calculate mean of non-outlier scores (Score <= 100)
mean_score = df5[df5['Score'] <= 100]['Score'].mean()

# Replace outliers with mean score
df5['Score'] = df5['Score'].apply(lambda x: mean_score if x > 100 else x)

print("DataFrame after replacing outliers with mean:")
print(df5)

#Part B (Image preprocessing)

from google.colab import files
# Opening a file picker to upload the image
uploaded = files.upload()

from PIL import Image
import matplotlib.pyplot as plt

#importing the image
image = Image.open("panda.jpg")

# Display image
plt.imshow(image)
plt.axis('off')
plt.show()

# Convert image to grayscale
gray_image = image.convert("L")

# Display grayscale image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()

#Resize image
resized_image = image.resize((1000, 500))
plt.imshow(resized_image)
plt.axis('off')
plt.title("Resized Image")
plt.show()

#Rotate the image
rotated_image = image.rotate(45) # Rotate 45 degrees
plt.imshow(rotated_image)
plt.axis('off')
plt.title("Rotated Image")
plt.show()

#flip the immage
flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(flipped_image)
plt.axis('off')
plt.title("Flipped Image")
plt.show()

#Crop a Region
# Crop box: (left, upper, right, lower)
cropped_image = image.crop((50, 50, 250, 250))
plt.imshow(cropped_image)
plt.axis('off')
plt.title("Cropped Image")
plt.show()

#Blurred Image
from PIL import ImageFilter
blurred = image.filter(ImageFilter.BLUR)
plt.imshow(blurred)
plt.axis('off')
plt.title("Blurred")
plt.show()

#Contour Filter
contour = image.filter(ImageFilter.CONTOUR)
plt.imshow(contour)
plt.axis('off')
plt.title("Contour")
plt.show()

#Detail Enhancement
detailed = image.filter(ImageFilter.DETAIL)
plt.imshow(detailed)
plt.axis('off')
plt.title("Detail Enhanced")
plt.show()

#Edge Enhancement
edges = image.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(edges)
plt.axis('off')
plt.title("Edge Enhanced")
plt.show()

#Emboss Filter
emboss = image.filter(ImageFilter.EMBOSS)
plt.imshow(emboss)
plt.axis('off')
plt.title("Embossed")
plt.show()

#Sharpened Filter
sharpened = image.filter(ImageFilter.SHARPEN)
plt.imshow(sharpened)
plt.axis('off')
plt.title("Sharpened")
plt.show()

#Custom Convolution Filter(Kernel)
# Example 3x3 sharpening kernel
kernel = ImageFilter.Kernel(
size=(3, 3),
kernel=[0, -1, 0, -1, 5, -1, 0, -1, 0],
scale=None, # Automatically calculated
offset=0
)
custom_filtered = image.filter(kernel)
plt.imshow(custom_filtered)
plt.axis('off')
plt.title("Custom Filtered")
plt.show()

#Edge Detection
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('panda.jpg', cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(img, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()

#Histogram Equalization
equalized = cv2.equalizeHist(img)
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalization")
plt.axis('off')
plt.show()

#Thresholding
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(binary, cmap='gray')
plt.title("Binary Threshold")
plt.axis('off')
plt.show()

#Color Space Conversion
color_img = cv2.imread('panda.jpg')
hsv = cv2.cvtColor(color_img, cv2.COLOR_BGR2HSV)
plt.imshow(cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB))
plt.title("HSV Converted")
plt.axis('off')
plt.show()


#Save image
gray_image.save("panda_gray.jpg")

#Download the image
from google.colab import files
files.download("panda_gray.jpg")






```
